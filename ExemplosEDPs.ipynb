```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, ifft
from scipy.integrate import solve_ivp
```

```python
# Configurações gerais para os gráficos
# Use a valid matplotlib style, 'seaborn-v0_8-poster' is a close alternative
plt.style.use('seaborn-v0_8-poster')
```


### 1. Equação do Calor (Heat Equation) ###

```python
def heat_equation_example():
    x = np.linspace(0, 1, 100)
    t = 0.1
    alpha = 0.01
    u = np.sin(np.pi * x) * np.exp(-np.pi**2 * alpha * t)
    plt.figure(figsize=(8, 4))
    plt.plot(x, u, label=f't = {t}')
    plt.title('Heat Equation: Temperature Distribution')
    plt.xlabel('Position (x)')
    plt.ylabel('Temperature (u)')
    plt.legend()
    plt.show()
```



### 2. Equação da Onda (Wave Equation) ###

```python
def wave_equation_example():
    x = np.linspace(0, 1, 1000)
    t = 0.05
    c = 1
    k = np.pi
    u = np.sin(k * x) * np.cos(k * c * t)
    plt.figure(figsize=(8, 4))
    plt.plot(x, u, label=f't = {t}')
    plt.title('Wave Equation: Displacement Over Time')
    plt.xlabel('Position (x)')
    plt.ylabel('Displacement (u)')
    plt.legend()
    plt.show()
```


### 3. Equação de Laplace (Laplace Equation) ###

```python
def laplace_equation_example():
    x = np.linspace(0, 1, 100)
    y = np.linspace(0, 1, 100)
    X, Y = np.meshgrid(x, y)
    U = np.sin(np.pi * X) * np.sin(np.pi * Y)
    plt.figure(figsize=(6, 5))
    plt.contourf(X, Y, U, 20, cmap='viridis')
    plt.colorbar()
    plt.title('Laplace Equation: 2D Potential Distribution')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```


### 4. Problemas de Valor de Contorno (Boundary Value Problems) ###

```python
def bvp_example():
    x = np.linspace(0, 1, 100)
    u = x * (1 - x)
    plt.figure(figsize=(8, 4))
    plt.plot(x, u, label='u(x) = x(1 - x)')
    plt.title('Boundary Value Problem: Simple Quadratic Shape')
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.legend()
    plt.show()
```


### 5. Aplicações da Transformada de Fourier (Fourier Transform Applications) ###

```python
def fourier_transform_example():
    N = 500
    T = 1.0 / 500.0
    x = np.linspace(0.0, N*T, N, endpoint=False)
    y = np.sin(50.0 * 2.0 * np.pi * x) + 0.5 * np.sin(80.0 * 2.0 * np.pi * x)
    yf = fft(y)
    xf = fftfreq(N, T)[:N//2]
    plt.figure(figsize=(8, 4))
    plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
    plt.title('Fourier Transform: Frequency Spectrum')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
# Executa todos os exemplos
heat_equation_example()
wave_equation_example()
laplace_equation_example()
bvp_example()
fourier_transform_example()
```
